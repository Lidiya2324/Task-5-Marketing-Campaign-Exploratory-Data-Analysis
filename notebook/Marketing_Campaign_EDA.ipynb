### Histograms

#### Income

```python
plt.figure(figsize=(8,5))

sns.histplot(df["income"], bins=30)

plt.title("Income Distribution")

plt.show()
```

The dataset represents most customers earn between 30,000 and 80,000 annually, which represents the main concentration of the distribution. The income pattern shows a slight right-skew, with fewer customers earning above 80,000. Only a very small number of customers fall into the high-income range (above 100,000), with rare outliers reaching around 150,000–160,000.

#### Age

```python
plt.figure(figsize=(8,5))

sns.histplot(df["age"], bins=25)

plt.title("Age Distribution")

plt.show()
```

```python
df["age"]=2025-df['year_birth']
```

The majority of customers are middle-aged adults, with most ages concentrated between approximately 45 and 65 years.

#### Boxplot

```python
plt.figure(figsize=(8,4))

sns.boxplot(x=df["income"])

plt.title("Income Boxplot")

plt.show()
```

Most income values fall within a reasonable range, while a few higher-income customers appear as outliers.

#### Scatter Plot

#### Income vs Wine Spending

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    x="income",
    y="mntwines",
    data=df
)

plt.title("Income vs Wine Spending")

plt.show()
```

Customers with higher incomes generally spend more on wine, although spending behavior varies across individuals.

#### Correlation Heatmap

```python
numeric=df.select_dtypes(include="number")
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    numeric.corr(),
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Heatmap")

plt.show()
```

Income and spending variables show positive correlations, especially across product categories. Spending channels are moderately correlated, while recency has weak relationships with most features.

#### Pairplot

```python
sns.pairplot(
    df[
        [
            "income",
            "mntwines",
            "mntmeatproducts",
            "mntgoldprods"
        ]
    ]
)

plt.show()
```

Customers who spend more on one product category generally also spend more on other product categories, indicating a positive relationship between spending across different product types. However, the relationships are moderate rather than very strong, and a few outliers are present.

#### Education Analysis

```python
plt.figure(figsize=(8,5))

sns.countplot(
    y="education",
    data=df
)

plt.title("Customer Education")

plt.show()
```

The majority of customers are graduates, with fewer customers holding Master's and PhD degrees, and only a small proportion having Basic education. This indicates that the dataset mainly consists of customers with higher educational attainment.

#### Marital Status

```python
plt.figure(figsize=(8,5))

sns.countplot(
    y="marital_status",
    data=df
)

plt.title("Marital Status Distribution")

plt.show()
```

Most customers are Married, followed by those who are Together and Single. Divorced and Widow customers represent the smallest groups in the dataset.

#### Top Spending Categories

```python
spending = df[
[
"mntwines",
"mntfruits",
"mntmeatproducts",
"mntfishproducts",
"mntsweetproducts",
"mntgoldprods"
]
].sum()

spending.plot(kind="bar")

plt.title("Total Spending by Product Category")

plt.show()
```

Customers spend the most on Wine, followed by Meat Products. Spending on Fruits, Sweets, and Gold Products is considerably lower, with Fish Products receiving slightly higher spending than Fruits and Sweets.
